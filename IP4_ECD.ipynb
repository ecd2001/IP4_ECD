3. Create an 1D array of 100 evenly spaced numbers between 0 and 100 using NumPy and save it to variable X.


```python
X = np.linspace(0,100,100)
```

    a. Create a plot using plt.subplots() and plot X versus X2


```python
X2 = np.square(X)

fig = fig,ax1=plt.subplots()

# set left y axis
ax1.set_ylabel('X', color='red', size='x-large')
ax1.tick_params(axis='y', labelcolor='red', labelsize='large')
ax1.plot(X, color='red')

# set right y axis
ax2=ax1.twinx()

ax2.set_ylabel('X squared', color='blue', size='x-large')
ax2.tick_params(axis='y', labelcolor='blue', labelsize='large')
ax2.plot(X2, color='blue')

plt.title('X verses X2')
plt.show()
```

    b. Create a scatter plot of X versus the exponential of X (np.exp(X))


```python
expo=np.exp(X)

plt.figure(figsize=(12,6))
plt.scatter(X, expo)
plt.xlabel('X')
plt.ylabel('Exponential of X')
plt.title('X verses the exponential of X')
plt.show()
```

5. Remove all the symbols (\$\,)from the 'Price' column and convert  and update it to  64-bit int numbers using astype(‘int64’) 

    Hint. Look for string functions like replace/strip). The data frame will be look like after that:
    

```python
car_sales['Price'] = car_sales['Price'].str.strip('$')
car_sales['Price'] = car_sales['Price'].str.replace(',','')
car_sales.head()
```

```python
car_sales['Price'] = car_sales['Price'].astype('int64')
car_sales.info()
```

6.	Add a column called 'Total Sales' to car_sales which cumulatively adds the 'Price' column look up the cumsum() function.


```python
car_sales['Total Sales'] = car_sales['Price'].cumsum(axis=0)
car_sales.head()
```

7.	Add a column called 'Sale Date' which lists a series of successive dates starting from today (your today). Check out “date_range” function and keep the periods=len(car_sales). The data frame will be look like that:

 


```python
car_sales['Sale Date']=pd.date_range(start='2/15/2023', periods=len(car_sales))
car_sales.head()
```

16.	Find out Pearson correlation between 'Total Sales','Price','Odometer (KM)'. Comment on the correlation 



```python
heat_map_corr = car_sales[['Total Sales','Price','Odometer (KM)']].corr(method='pearson')
print(heat_map_corr)

cols=['Total Sales', 'Price', 'Odometer(KM)']

heat_map=sns.heatmap(heat_map_corr,
                    annot=True,
                    xticklabels=cols,
                    yticklabels=cols)
```
